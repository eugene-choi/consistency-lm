```python
lstm_results = {}

out_files = os.listdir(os.getcwd())

for out_file in out_files:
    print(out_file)
    file_info = out_file.split('_')
    if ".out" in file_info[1]:
        alpha = float(file_info[1].replace(".out",""))
        n_layers = 2
    else:
        alpha = float(file_info[1])
        n_layers = int(file_info[2].replace(".out",""))

    with open(out_file, mode="r") as file:
        for line in file.readlines():
            if 'best' in line:
                best_ppl = line
        
        ppl = best_ppl.split(" ")[-1].replace(")\n","")        
        # print(f'ppl = {ppl}')
        
        file.seek(0) # move the cursor of the filereader to the start of the file.
        find_epoch_info = True
        file = list(file.readlines())
        for idx, line in enumerate(file):
            if find_epoch_info and "Step:  2400" in line:
                epoch_info = line
            if "PPL "+ppl in line:
                best_ppl_line = line
                best_ppl_length = file[idx+1]
                find_epoch_info = False
        decoding_stats = ast.literal_eval(''.join(file[-18:]))
        # print(f'epoch_info=\n{epoch_info}')
        # epoch_info = re.split(r'[\s,]+', epoch_info)
        nll_loss, os_rate = None, None
        
        epoch_info = re.split(r'[\s,]+', epoch_info)
        for idx, token in enumerate(epoch_info):
            # if nll_loss not None and token == "NLL" and epoch_info[idx+1] == "loss:":
                # nll_loss = epoch_info[idx+2]
            if token == "OS" and epoch_info[idx+1] == "rate:": 
                osr = epoch_info[idx+2]
        
        best_ppl_line = re.split(r'[\s,]+', best_ppl_line)
        for idx, token in enumerate(best_ppl_line):
            if token == "NLL" and best_ppl_line[idx+1] == "loss":
                nll_loss = best_ppl_line[idx+2]
            # elif os_rate not None token == "OS" and epoch_info[idx+1] == "rate:": 
                # osr = epoch_info[idx+2]
        # print(best_ppl_line)
        
        # print(f'nll_loss = {nll_loss}')
        # print(f'osr = {osr}')
        
        
        # print(best_ppl_length)
        best_ppl_length = re.split(r'[\s,]+', best_ppl_length.replace(")",""))
        def get_float(x):
            try:
                return float(x)
            except ValueError:
                pass

        nums = []
        for x in list(filter(lambda _: _ is not None, map(get_float, best_ppl_length))):
            nums.append(x)
        nonterminated, avg_len, uniq_nonterminated = nums
        train_stats = {'ppl':float(ppl),
                       'nll':float(nll_loss),
                       'osr':float(osr),
                       'nonterminated':float(nonterminated),
                       'avg_len':float(avg_len),
                       'uniq_nonterminated':float(uniq_nonterminated),
                      }
        
        lstm_results.setdefault(int(n_layers),{})
        lstm_results[int(n_layers)].setdefault(float(alpha), {
            'train_stats':train_stats,
            'decoding_stats':decoding_stats})
        
```

```python
def get_plot(n_layer,key,xlabel):
    alphas = sorted(list(lstm_results[n_layer].keys()))
    return_dict = {}
    
    for alpha in alphas:
        cur_key = f'lstm_results[{n_layer}][{alpha}]'+key
        return_dict[alpha] = eval(cur_key)
        
    fig = plt.figure(figsize=(15,9))
    plt.plot(*zip(*sorted(return_dict.items())),linestyle='--', marker='o', color='b', label='line with marker')
    plt.title(f'{n_layer} layer LSTM from scratch on wikitext-2 w/ OSL', fontsize=20)
    plt.xlabel("α", fontsize=15)
    plt.ylabel(xlabel, fontsize=15)
    fig.savefig(f'{xlabel}.png', transparent=False)
    plt.show()
```
